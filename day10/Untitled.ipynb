```python
def find_s(contents):
    for i in range(len(contents)):
        if 'S' in contents[i]:
            return (i, contents[i].index('S'))
```

```python
import sys
sys.setrecursionlimit(1000000)
```

```python
def populate_distance(contents, distances, point, last_point):
    if distances[point[0]][point[1]] != -1:
        return distances[point[0]][point[1]]
    
    if contents[point[0]][point[1]] == 'S':
        distances[point[0]][point[1]] = 0
        neighbours = []
        # check left is '-' or 'L' or 'F'
        # check right is '-' or 'J' or '7'
        if point[1] > 0:
            left = contents[point[0]][point[1] - 1]
            if left in ['-', 'L', 'F']:
                neighbours.append((point[0], point[1] - 1))
        if point[1] < len(contents[0]) - 1:
            right = contents[point[0]][point[1] + 1]
            if right in ['-', 'J', '7']:
                neighbours.append((point[0], point[1] + 1))
        
        # check top is '|' or '7' or 'F'
        # check bottom is '|' or 'L' or 'J'
        if point[0] > 0:
            top = contents[point[0] - 1][point[1]]
            if top in ['|', '7', 'F']:
                neighbours.append((point[0] - 1, point[1]))
                
        if point[0] < len(contents) - 1:
            bottom = contents[point[0] + 1][point[1]]
            if bottom in ['|', 'L', 'J']:
                neighbours.append((point[0] + 1, point[1]))
                
        return max(populate_distance(contents, distances, neighbours[0], point), populate_distance(contents, distances, neighbours[1], point))
    
    current_distance = distances[last_point[0]][last_point[1]] + 1
    distances[point[0]][point[1]] = current_distance
    next_distance = 0
    
    if contents[point[0]][point[1]] == '|':
        if point[0] > last_point[0]:
            next_distance = populate_distance(contents, distances, (point[0] + 1, point[1]), point)
        else:
            next_distance = populate_distance(contents, distances, (point[0] - 1, point[1]), point)
            
    if contents[point[0]][point[1]] == '-':
        if point[1] > last_point[1]:
            next_distance = populate_distance(contents, distances, (point[0], point[1] + 1), point)
        else:
            next_distance = populate_distance(contents, distances, (point[0], point[1] - 1), point)
            
    if contents[point[0]][point[1]] == 'L':
        if point[0] == last_point[0]:
            next_distance = populate_distance(contents, distances, (point[0] - 1, point[1]), point)
        else:
            next_distance = populate_distance(contents, distances, (point[0], point[1] + 1), point)

    if contents[point[0]][point[1]] == 'J':
        if point[0] == last_point[0]:
            next_distance = populate_distance(contents, distances, (point[0] - 1, point[1]), point)
        else:
            next_distance = populate_distance(contents, distances, (point[0], point[1] - 1), point)
        
    if contents[point[0]][point[1]] == '7':
        if point[0] == last_point[0]:
            next_distance = populate_distance(contents, distances, (point[0] + 1, point[1]), point)
        else:
            next_distance = populate_distance(contents, distances, (point[0], point[1] - 1), point)
    
    if contents[point[0]][point[1]] == 'F':
        if point[0] == last_point[0]:
            next_distance = populate_distance(contents, distances, (point[0] + 1, point[1]), point)
        else:
            next_distance = populate_distance(contents, distances, (point[0], point[1] + 1), point)
    
    return max(current_distance, next_distance)
    
    
```

```python
with open('input.txt') as file:
    contents = file.readlines()
    contents = [line.strip() for line in contents]
    s_location = find_s(contents)
    distances = [None] * len(contents)
    for i in range(len(contents)):
        distances[i] = [-1] * len(contents[i])
    result = populate_distance(contents, distances, s_location, None)
```

```python
def is_in_polygon(contents, point):
    counts = 0
    for i in range(point[1]):
        if contents[point[0]][i] != '.' and contents[point[0]][i] not in ['-', 'F', '7']:
            counts += 1
    return counts % 2 == 1
```
