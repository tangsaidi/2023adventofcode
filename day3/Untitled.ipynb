```python
def parse_number(contents, x, y):
    while y > 0:
        if contents[x][y-1].isnumeric():
            y -= 1
        else:
            break
    begin = y
    while y < len(contents) - 1:
        if contents[x][y + 1].isnumeric():
            y += 1
        else:
            break
    end = y
    return int(contents[x][begin:end+1])
```

```python
def has_exactly_two_adjacent_numbers(contents, x, y):
    numbers_count = 0
    parsed_numbers = []
    if y > 0 and contents[x][y-1].isnumeric(): # left
        numbers_count += 1
        parsed_numbers.append(parse_number(contents, x, y - 1))
    if y < len(contents[x]) - 1 and contents[x][y+1].isnumeric(): # right
        numbers_count += 1
        parsed_numbers.append(parse_number(contents, x, y + 1))
        
    if x > 0 and contents[x-1][y].isnumeric():
        numbers_count += 1
        parsed_numbers.append(parse_number(contents, x-1, y))
    else:
        if y > 0 and contents[x-1][y-1].isnumeric():
            numbers_count += 1
            parsed_numbers.append(parse_number(contents, x-1, y-1))
        if y < len(contents[x-1]) - 1 and contents[x-1][y+1].isnumeric():
            numbers_count += 1
            parsed_numbers.append(parse_number(contents, x-1, y+1))
    
    if x < len(contents) - 1 and contents[x+1][y].isnumeric():
        numbers_count += 1
        parsed_numbers.append(parse_number(contents, x+1, y))
    else:
        if y > 0 and contents[x+1][y-1].isnumeric():
            numbers_count += 1
            parsed_numbers.append(parse_number(contents, x+1, y-1))
        if y < len(contents[x+1]) - 1 and contents[x+1][y+1].isnumeric():
            numbers_count += 1
            parsed_numbers.append(parse_number(contents, x+1, y+1))
    if numbers_count == 2:
        return (parsed_numbers[0], parsed_numbers[1])
    return None
```
