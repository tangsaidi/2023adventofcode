```python
with open('input.txt') as file:
    contents = file.readlines()
    contents = [line.strip() for line in contents]
    instructions = contents[0]
    mappings = contents[2:]
    mappings_dict = {}
    for mapping in mappings:
        processed = mapping.split(' = ')
        key = processed[0]
        value = processed[1]
        value = value[1:-1]
        value = value.split(', ')
        mappings_dict[key] = value
    
    seeds = list(filter(lambda x: x[-1] == 'A', list(mappings_dict.keys())))
    
    seed_set= set()
    seed_set.add(''.join(seeds))
    steps = 0
    instruction_length = len(instructions)
    while True:
        cur_instruction = steps % instruction_length
        for i in range(len(seeds)):        
            if instructions[cur_instruction] == 'L':
                seeds[i] = mappings_dict[seeds[i]][0]
            else:
                seeds[i] = mappings_dict[seeds[i]][1]
        steps += 1
        # print(seeds)
        all_z = True
        for seed in seeds:
            if seed[-1] != 'Z':
                all_z = False
                break
        if all_z:
            break
        else:
            if ''.join(seeds) in seed_set:
                break
            else:
                seed_set.add(''.join(seeds))
            if len(seed_set) % 10_000_000 == 0:
                print("tested seed: ", len(seed_set))

        # left = value[0]
        # right = value[1]
        # print("key: ", key, " value: ", value)
```
